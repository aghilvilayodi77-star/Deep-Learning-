word2vec

```python
import string
import nltk
from nltk.corpus import brown
from gensim.models import Word2Vec
from sklearn.decomposition import PCA
from matplotlib import pyplot as plt
```

```python
nltk.download("brown")
```

```python
document=brown.sents()
data=[]
for sents in document:
  new_sents=[]
  for word in sents:
    new_word=word.lower()
    if new_word not in string.punctuation:
      new_sents.append(new_word)
      if new_sents:
         data.append(new_sents)
```

```python
#creating word2vect
model=Word2Vec(
    sentences=data,
    vector_size=50,
    window=10,
    epochs=20,

)
```

```python
print("vector for 'like':")
print(model.wv['like'])
print()
```

```python
print("Three word similar to 'car':")
if "car" in model.wv:
  words=model.wv.most_similar("car",topn=3)
  for word in words:
    print(word)
else:
  print("The word 'car' is not in the vocabulary")
  print()
```

BAg of word (Bow) convert sentenece into numerical vector

```python
from sklearn.feature_extraction.text import CountVectorizer
documents=["we love cats","cat are great pets"]
vectorizer=CountVectorizer(lowercase=True,stop_words=None)
bow=vectorizer.fit_transform(documents)
print("Bag.of.word matrix:")
print(bow.toarray())
print()
print("Vocabulary:(features)")
print(vectorizer.get_feature_names_out())
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
documents=['i love cat','cat are great pets']
vect=TfidfVectorizer()
tfidf=vect.fit_transform(documents)
tfidf_array=tfidf.toarray()
vocabulary=vect.get_feature_names_out()
(tfidf_array,vocabulary)
```
